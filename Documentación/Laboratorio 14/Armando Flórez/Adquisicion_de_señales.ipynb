```python
import csv

# Nombre del archivo de entrada y salida
archivo_txt = 'ECG_reposo.txt'
archivo_csv = 'ecg_reposo.csv'
archivo_csv_recortado = 'señal_recortada_ecg_reposo.csv'

# Abrir el archivo de entrada
with open(archivo_txt, 'r') as txt_file:
    # Leer todas las líneas del archivo
    lineas = txt_file.readlines()

# Saltear las primeras 2 líneas
lineas = lineas[3:]

# Limitar a las primeras 500 filas
lineas = lineas[:500]

# Crear una lista para las filas del archivo CSV
filas_csv = []
filas_csv_recortado = []

# Añadir los títulos de las columnas
filas_csv.append(['timestamp', 'signal'])
filas_csv_recortado.append(['timestamp', 'signal'])

# Inicializar contador para la primera columna del archivo CSV
contador = 1

# Procesar cada línea restante del archivo
for linea in lineas:
    # Dividir la línea en columnas
    columnas = linea.strip().split()
    # Obtener solo las columnas necesarias (numeración y sexta columna)
    nueva_fila = [contador, columnas[5]]
    # Agregar la nueva fila a la lista de filas
    filas_csv.append(nueva_fila)
    filas_csv_recortado.append(nueva_fila)
    # Incrementar el contador
    contador += 1

# Escribir las filas en el archivo CSV completo
with open(archivo_csv, 'w', newline='') as csv_file:
    escritor_csv = csv.writer(csv_file)
    # Escribir las filas
    escritor_csv.writerows(filas_csv)

# Escribir las filas en el archivo CSV recortado
with open(archivo_csv_recortado, 'w', newline='') as csv_file_recortado:
    escritor_csv_recortado = csv.writer(csv_file_recortado)
    # Escribir las filas
    escritor_csv_recortado.writerows(filas_csv_recortado)

print(f'Archivo CSV generado: {archivo_csv}')
print(f'Archivo CSV recortado generado: {archivo_csv_recortado}')
```
