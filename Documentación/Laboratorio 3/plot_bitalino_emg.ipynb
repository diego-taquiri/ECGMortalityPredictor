```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("emg_raw_data/isb-reposo-armando/opensignals_98D3B1FD3DA9_2024-04-12_12-12-05.txt",skiprows=3)
data_mV = data[:,5]/1024 *3.223 - 3.223/2 #De acuerdo a la cuantizacin de 3.223 mV a 1024 bits del manual de Bitalino
time = np.arange(len(data_mV))/1000 

plt.figure(figsize=(20,12))

plt.plot(time,data_mV)

plt.ylim(-2, 2)
plt.xlim(13, 31)
plt.tick_params(axis='both', which='major', labelsize=34, width=2, length=10)
plt.xlabel('Tiempo (s)', fontsize=45)
plt.ylabel('mV', fontsize=45)
plt.title('EMG reposo persona 1', fontsize=45)
plt.tight_layout()

plt.savefig("plots/isb-reposo-armando.png")
```
